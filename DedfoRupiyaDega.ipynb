```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

data = pd.read_csv('train_FD001.csv',header=None, names=column_names)

#Defined a function to calculate the remaining useful life (RUL)
def add_rul_column(df):
    df['RUL'] = df.groupby('Unit')['Cycle'].transform(max) - df['Cycle']
    return df

#Adding the RUL column to the dataset
data = add_rul_column(data)

#The features except non-sensor values
features = data.drop(columns=['RUL', 'Unit', 'Cycle', 'Op_Setting_1', 'Op_Setting_2', 'Op_Setting_3'])  # Drop non-sensor columns
target = data['RUL']

kf = KFold(n_splits=4, shuffle=True, random_state=42)

#Defining the List of RMSE
rmse_values = []

#looping eacch fold
for train_index, test_index in kf.split(data['Unit'].unique()):
    train_Units = data['Unit'].unique()[train_index]
    test_Units = data['Unit'].unique()[test_index]

    #Split the dataset into training and validation sets
    train_data = data[data['Unit'].isin(train_Units)]
    test_data = data[data['Unit'].isin(test_Units)]

    #For the test set, restrict the data to 70% of the life Cycle for each engine
    test_data_70_percent = test_data.groupby('Unit').apply(lambda x: x.iloc[:int(0.7 * len(x))]).reset_index(drop=True)

    #Train data definition
    X_train = train_data.drop(columns=['RUL', 'Cycle'])
    y_train = train_data['RUL']


    X_test = test_data_70_percent.drop(columns=['RUL', 'Cycle'])
    y_test_actual = test_data_70_percent.groupby('Unit').apply(lambda x: x['RUL'].iloc[-1]).reset_index(drop=True)
    #Training ya Fitting linear regression model
    lr = LinearRegression()
    lr.fit(X_train, y_train)

    #Predict remaining cycles for the validation set
    y_pred = lr.predict(X_test.groupby('Unit').tail(1))  # Predict at the 70% Cycle point

    #The Error e = A - (N - P)
    N_P = test_data.groupby('Unit')['RUL'].max().values - y_test_actual.values
    error = y_pred - N_P

    #Calculating RMSE
    rmse = np.sqrt(mean_squared_error(N_P, y_pred))
    rmse_values.append(rmse)

    #Plot actual vs predicted RUL
    plt.scatter(N_P, y_pred, label=f'Fold {len(rmse_values)}')
    plt.xlabel('Actual Remaining Cycles')
    plt.ylabel('Predicted Remaining Cycles')
    plt.legend()
    plt.savefig('LR.png')

#Show plot
plt.title('Actual vs Predicted Remaining Cycles')
plt.show()


#Printing RMSE's
for i, rmse in enumerate(rmse_values):
    print(f"Fold {i+1} RMSE: {rmse}")

#A summary
print(f"Mean RMSE across folds: {np.mean(rmse_values)}")
```
